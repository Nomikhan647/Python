# Series create, manipulate, querry, delete

```python
#creating a series from a list
arr = [1,2,3,4,5]
s1 = pd.Series(arr)
s1
```

```python
order = [1,2,3,4,5]
s2 = pd.Series(arr, index=order)
s2
```

```python
import numpy as np
n = np.random.randn(5)
index = ['a','b','c','d','e']
s = pd.Series(n, index=index)
s
```

```python
#create series from dictionary
d = {'a':1, 'b':2, 'c':3}
s3 = pd.Series(d)
s3
```

```python
s3.index = [1,2,3]
s3
```

```python
#slicing
s3[:3]
```
